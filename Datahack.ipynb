```python
import numpy as np
import pandas as pd
```

```python
x_train = pd.read_csv('/content/training_set_features.csv')
y_train = pd.read_csv('/content/training_set_labels.csv')
x_test = pd.read_csv('/content/test_set_features.csv')
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
df = pd.merge(x_train, y_train, on='respondent_id')
```

```python
(df.isnull().sum()/len(df))*100
```

```python
(x_test.isnull().sum()/len(x_test))*100
```

```python
df.drop(columns = ['employment_occupation','employment_industry'], inplace = True)
x_test.drop(columns = ['employment_occupation','employment_industry'], inplace = True)
```

```python
for i in df.columns[1:36]:
  print(df[i].value_counts())
```

```python
df['education'].replace({'< 12 Years': 0, '12 Years': 1, 'Some College': 2, 'College Graduate': 3}, inplace = True)

x_test['education'].replace({'< 12 Years': 0, '12 Years': 1, 'Some College': 2, 'College Graduate': 3}, inplace = True)

df['race'].replace({'White': 0, 'Black': 1, 'Hispanic': 2, 'Other or Multiple': 3}, inplace = True)

x_test['race'].replace({'White': 0, 'Black': 1, 'Hispanic': 2, 'Other or Multiple': 3}, inplace = True)

df['sex'].replace({'Female': 1, 'Male': 0}, inplace = True)

x_test['sex'].replace({'Female': 1, 'Male': 0}, inplace = True)

df['income_poverty'].replace({'<= $75,000, Above Poverty': 1, '> $75,000': 2, 'Below Poverty': 0}, inplace = True)

x_test['income_poverty'].replace({'<= $75,000, Above Poverty': 1, '> $75,000': 2, 'Below Poverty': 0}, inplace = True)

df['marital_status'].replace({'Married': 1, 'Not Married': 0}, inplace = True)

x_test['marital_status'].replace({'Married': 1, 'Not Married': 0}, inplace = True)

df['rent_or_own'].replace({'Rent': 0, 'Own': 1}, inplace = True)

x_test['rent_or_own'].replace({'Rent': 0, 'Own': 1}, inplace = True)

df['employment_status'].replace({'Unemployed': 0, 'Employed': 1, 'Not in Labor Force': 2}, inplace = True)

x_test['employment_status'].replace({'Unemployed': 0, 'Employed': 1, 'Not in Labor Force': 2}, inplace = True)

df['census_msa'].replace({'MSA, Not Principle  City': 2, 'MSA, Principle City': 1, 'Non-MSA': 0}, inplace = True)

x_test['census_msa'].replace({'MSA, Not Principle  City': 2, 'MSA, Principle City': 1, 'Non-MSA': 0}, inplace = True)

df['hhs_geo_region'].replace({'lzgpxyit': 0, 'fpwskwrf': 1, 'qufhixun': 2, 'oxchjgsf': 3, 'kbazzjca': 4, 'bhuqouqj': 5, 'mlyzmhmf': 6, 'lrircsnp': 7, 'atmpeygn': 8, 'dqpwygqj': 9}, inplace = True)

x_test['hhs_geo_region'].replace({'lzgpxyit': 0, 'fpwskwrf': 1, 'qufhixun': 2, 'oxchjgsf': 3, 'kbazzjca': 4, 'bhuqouqj': 5, 'mlyzmhmf': 6, 'lrircsnp': 7, 'atmpeygn': 8, 'dqpwygqj': 9}, inplace = True)
```

```python
df['age_group'].replace({'65+ Years': 4, '55 - 64 Years': 3, '45 - 54 Years': 2, '35 - 44 Years': 1, '18 - 34 Years': 0}, inplace  = True)

x_test['age_group'].replace({'65+ Years': 4, '55 - 64 Years': 3, '45 - 54 Years': 2, '35 - 44 Years': 1, '18 - 34 Years': 0}, inplace  = True)
```

```python
for i in df.columns[1:36]:
  print(df[i].value_counts())
```

```python
for i in x_test.columns[1:36]:
  print(x_test[i].value_counts())
```

```python
df.drop(columns = ['health_insurance'], inplace = True)
x_test.drop(columns = ['health_insurance'], inplace = True)
```

```python
df.info()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'most_frequent')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
x_test_imputed = pd.DataFrame(imputer.fit_transform(x_test), columns=x_test.columns)
```

```python
df_imputed.info()
```

```python
from sklearn.model_selection import train_test_split
x_train_xyz, x_val_xyz, y_train_xyz, y_val_xyz = train_test_split(df_imputed.iloc[:, 1:33], df_imputed['xyz_vaccine'], test_size=0.2, random_state=42)
x_train_seasonal, x_val_seasonal, y_train_seasonal, y_val_seasonal = train_test_split(df_imputed.iloc[:, 1:33], df_imputed['seasonal_vaccine'], test_size=0.2, random_state=42)
```

```python
x_test = x_test_imputed.iloc[:, 1:]
```

```python
from sklearn.linear_model import LogisticRegression
lr1 = LogisticRegression()
lr2 = LogisticRegression()
lr1.fit(x_train_xyz, y_train_xyz)
lr2.fit(x_train_seasonal,y_train_seasonal)
```

```python
lr1.classes_
```

```python
y_pred_xyz_lr1 = lr1.predict_proba(x_val_xyz)[:,1]
y_pred_seasonal_lr2 = lr2.predict_proba(x_val_seasonal)[:,1]
```

```python
print(y_pred_xyz_lr1)
```

```python
print(y_pred_seasonal_lr2)
```

```python
from sklearn.naive_bayes import CategoricalNB
nb1 = CategoricalNB()
nb2 = CategoricalNB()
nb1.fit(x_train_xyz, y_train_xyz)
nb2.fit(x_train_seasonal, y_train_seasonal)
```

```python
nb1.classes_
```

```python
y_pred_xyz_nb1 = nb1.predict_proba(x_val_xyz)[:,1]
y_pred_seasonal_nb2 = nb2.predict_proba(x_val_seasonal)[:,1]
```

```python
print(y_pred_xyz_nb1)
```

```python
print(y_pred_seasonal_nb2)
```

```python
from sklearn.svm import SVC
svc1 = SVC(probability = True, C= 0.1, kernel = 'linear', gamma = 'scale')
svc2 = SVC(probability = True, C= 0.1, kernel = 'linear', gamma = 'scale')
svc1.fit(x_train_xyz, y_train_xyz)
svc2.fit(x_train_seasonal, y_train_seasonal)
```

```python
svc1.classes_
```

```python
y_pred_xyz_svc1 = svc1.predict_proba(x_val_xyz)[:,1]
y_pred_seasonal_svc2 = svc2.predict_proba(x_val_seasonal)[:,1]
```

```python
print(y_pred_xyz_svc1)
```

```python
print(y_pred_seasonal_svc2)
```

```python
param_grid = {
    'C': [0.01, 0.1, 1.0, 10.0, 100.0],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}
```

```python
from sklearn.model_selection import GridSearchCV
model = LogisticRegression()
grid_search1 = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)
grid_search2 = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)

grid_search1.fit(x_train_xyz, y_train_xyz)
grid_search2.fit(x_train_seasonal, y_train_seasonal)

print(f"Best parameters: {grid_search1.best_params_}")
print(f"Best parameters: {grid_search2.best_params_}")

best_model1 = grid_search1.best_estimator_
best_model2 = grid_search2.best_estimator_
```

```python
best_model1.fit(x_train_xyz, y_train_xyz)
best_model2.fit(x_train_seasonal, y_train_seasonal)
```

```python
best_model1.classes_
```

```python
y_pred_xyz_lrh1 = best_model1.predict_proba(x_val_xyz)[:,1]
y_pred_seasonal_lrh2 = best_model2.predict_proba(x_val_seasonal)[:,1]
```

```python
print(y_pred_xyz_lrh1)
```

```python
print(y_pred_seasonal_lrh2)
```

```python
param_grid = {
    'alpha': [0.1, 0.5, 1.0, 2.0]
}
```

```python
cnb = CategoricalNB()
grid_search1 = GridSearchCV(estimator=cnb, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)
grid_search2 = GridSearchCV(estimator=cnb, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)

grid_search1.fit(x_train_xyz, y_train_xyz)
grid_search2.fit(x_train_seasonal, y_train_seasonal)

print(f"Best parameters: {grid_search1.best_params_}")
print(f"Best parameters: {grid_search2.best_params_}")

best_cnb1 = grid_search1.best_estimator_
best_cnb2 = grid_search2.best_estimator_
```

```python
best_cnb1.classes_
```

```python
y_pred_xyz_nbh1 = best_cnb1.predict_proba(x_val_xyz)[:,1]
y_pred_seasonal_nbh2 = best_cnb2.predict_proba(x_val_seasonal)[:,1]
```

```python
# Calculate AUC-ROC scores
from sklearn.metrics import roc_auc_score, roc_curve
lr1_auc = roc_auc_score(y_val_xyz, y_pred_xyz_lr1)
svc1_auc = roc_auc_score(y_val_xyz, y_pred_xyz_svc1)
nb1_auc = roc_auc_score(y_val_xyz, y_pred_xyz_nb1)
lrh1_auc = roc_auc_score(y_val_xyz, y_pred_xyz_lrh1)
nbh1_auc = roc_auc_score(y_val_xyz, y_pred_xyz_nbh1)

# Calculate AUC-ROC scores
lr2_auc = roc_auc_score(y_val_seasonal, y_pred_seasonal_lr2)
svc2_auc = roc_auc_score(y_val_seasonal, y_pred_seasonal_svc2)
nb2_auc = roc_auc_score(y_val_seasonal, y_pred_seasonal_nb2)
lrh2_auc = roc_auc_score(y_val_seasonal, y_pred_seasonal_lrh2)
nbh2_auc = roc_auc_score(y_val_seasonal, y_pred_seasonal_nbh2)
```

```python
# Print AUC-ROC scores
print(f"Logistic Regression1 AUC-ROC: {lr1_auc}")
print(f"SVM1 AUC-ROC: {svc1_auc}")
print(f"Naive Bayes1 AUC-ROC: {nb1_auc}")
print(f"Logistic Regression with Hyperparameter tuning1 AUC-ROC: {lrh1_auc}")
print(f"Naive Bayes with Hyperparameter tuning1 AUC-ROC: {nbh1_auc}")

print(f"Logistic Regression2 AUC-ROC: {lr2_auc}")
print(f"SVM2 AUC-ROC: {svc2_auc}")
print(f"Naive Bayes2 AUC-ROC: {nb2_auc}")
print(f"Logistic Regression with Hyperparameter tuning2 AUC-ROC: {lrh2_auc}")
print(f"Naive Bayes with Hyperparameter tuning2 AUC-ROC: {nbh2_auc}")
```

```python
#Plot AUC-ROC curve
lr1_fpr, lr1_tpr, _ = roc_curve(y_val_xyz, y_pred_xyz_lr1)
svc1_fpr, svc1_tpr, _ = roc_curve(y_val_xyz, y_pred_xyz_svc1)
nb1_fpr, nb1_tpr, _ = roc_curve(y_val_xyz, y_pred_xyz_nb1)
lrh1_fpr, lrh1_tpr, _ = roc_curve(y_val_xyz, y_pred_xyz_lrh1)
nbh1_fpr, nbh1_tpr, _ = roc_curve(y_val_xyz, y_pred_xyz_nbh1)

#Plot AUC-ROC curve
lr2_fpr, lr2_tpr, _ = roc_curve(y_val_seasonal, y_pred_seasonal_lr2)
svc2_fpr, svc2_tpr, _ = roc_curve(y_val_seasonal, y_pred_seasonal_svc2)
nb2_fpr, nb2_tpr, _ = roc_curve(y_val_seasonal, y_pred_seasonal_nb2)
lrh2_fpr, lrh2_tpr, _ = roc_curve(y_val_seasonal, y_pred_seasonal_lrh2)
nbh2_fpr, nbh2_tpr, _ = roc_curve(y_val_seasonal, y_pred_seasonal_nbh2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(lr1_fpr, lr1_tpr, label=f"Logistic Regression1 (AUC = {lr1_auc:.2f})")
plt.plot(svc1_fpr, svc1_tpr, label=f"SVM1 (AUC = {svc1_auc:.2f})")
plt.plot(nb1_fpr, nb1_tpr, label=f"Naive Bayes1 (AUC = {nb1_auc:.2f})")
plt.plot(lrh1_fpr, lrh1_tpr, label=f"Logistic Regression with HT 1 (AUC = {lrh1_auc:.2f})")
plt.plot(nbh1_fpr, nbh1_tpr, label=f"Naive Bayes with HT 1 (AUC = {nbh1_auc:.2f})")

plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(lr2_fpr, lr2_tpr, label=f"Logistic Regression2 (AUC = {lr2_auc:.2f})")
plt.plot(svc2_fpr, svc2_tpr, label=f"SVM2 (AUC = {svc2_auc:.2f})")
plt.plot(nb2_fpr, nb2_tpr, label=f"Naive Bayes2 (AUC = {nb2_auc:.2f})")
plt.plot(lrh2_fpr, lrh2_tpr, label=f"Logistic Regression with HT 2 (AUC = {lrh2_auc:.2f})")
plt.plot(nbh2_fpr, nbh2_tpr, label=f"Naive Bayes with HT 2 (AUC = {nbh2_auc:.2f})")

plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.show()
```
